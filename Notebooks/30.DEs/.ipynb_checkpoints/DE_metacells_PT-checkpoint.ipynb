```r
seurat_obj@meta.data$total_cells = rowSums(
  cbind(
    seurat_obj@meta.data$X101010_total_cells,
    seurat_obj@meta.data$SEAAD_total_cells,
    seurat_obj@meta.data$Gaz_total_cells,
    seurat_obj@meta.data$Mathys_total_cells
  ),
  na.rm = TRUE
)
```

```r
seurat_obj@meta.data$pct_SEAAD <- ifelse(is.na(seurat_obj@meta.data$SEAAD_total_cells), 0, 
  seurat_obj@meta.data$SEAAD_total_cells) / seurat_obj@meta.data$total_cells

seurat_obj@meta.data$pct_101010 <- ifelse(is.na(seurat_obj@meta.data$X101010_total_cells), 0, 
  seurat_obj@meta.data$X101010_total_cells) / seurat_obj@meta.data$total_cells

seurat_obj@meta.data$pct_Gaz <- ifelse(is.na(seurat_obj@meta.data$Gaz_total_cells), 0, 
  seurat_obj@meta.data$Gaz_total_cells) / seurat_obj@meta.data$total_cells

seurat_obj@meta.data$pct_Mathys <- ifelse(is.na(seurat_obj@meta.data$Mathys_total_cells), 0, 
  seurat_obj@meta.data$Mathys_total_cells) / seurat_obj@meta.data$total_cells
```

```r

# prepare data2 for glmmTMB
# Extract the expression matrix and metadata2
expr_matrix <- as.data.frame(t(as.matrix(GetAssayData(seurat_obj, layer = "counts",assay="RNA"))))
meta_data <- seurat_obj@meta.data

new_genenames <- gsub("-", "", colnames(expr_matrix))
colnames(expr_matrix) <- new_genenames


# Filter out low exp genes 
# keep if more than 50 raw counts are detected in at least 5% of cells
gene_expression_counts <- colSums(expr_matrix > 50)
total_cells <- nrow(expr_matrix)
percent_expressed <- (gene_expression_counts / total_cells) * 100
genes_to_keep <- ((colnames(expr_matrix))[which(((gene_expression_counts / total_cells) * 100 >= 5))])

print(length(genes_to_keep))

# subset object
expr_matrix <- expr_matrix[,genes_to_keep]

# Combine expression data2 with metadata2
data2 <- cbind(meta_data, expr_matrix)


```

```r
options(repr.plot.width = 18, repr.plot.height = 15)
library(ggrepel)

cbPalette <- c("#3B4CC0", "grey", "#B40426", "#F0E442", "#0072B2", "#D55E00", "#CC79A7")

volPlot_nebula <- function(markers, colors) {
  
  # Add gene label column
  markers$X <- markers$Gene
  
  # Define significance and direction for colouring
  markers$direction <- ifelse(markers$LogFC < 0 & markers$p_adj < 0.05, "Down-regulated",
                       ifelse(markers$LogFC > 0 & markers$p_adj < 0.05, "Up-regulated", "NS"))
  
  # Define which genes to label
  markers$label <- ifelse(
    (markers$p_adj < 0.05 & abs(markers$LogFC) > 0.0001) | (abs(markers$LogFC) > 10),
    markers$X, ""
  )
  
  ggplot(markers, aes(x = LogFC, y = -log10(p_adj))) +
    
    # pts
    geom_point(
      aes(fill = factor(direction)),
      size = 4, shape = 21, stroke = 0.1,
      color = "grey2", alpha = 0.8,
      key_glyph = draw_key_point
    ) +
    
    #  labels
    geom_label_repel(
      aes(label = label),
      fill  = alpha("white", 0.1),
      color = "black",
      size  = 5,
      max.overlaps = 40,
      force = 0.5
    ) +
    
    # Reference lines
    geom_hline(yintercept = 1.3,  color = "red",   linetype = "dashed") +
    geom_vline(xintercept = 0,    color = "black",  linetype = "dashed") +
    
    scale_fill_manual(
      name   = "Direction (padj < 0.05)",
      values = colors
    ) +
    scale_color_manual(
      name   = "Gene Set",
      values = c("red", "blue", NA, NA)
    ) +
    scale_alpha_manual(values = c(0.8, 0.8, 0, 0), guide = "none") +
    
    xlab("\navg_log2FoldChange") +
    ylab("-log10(padj)\n") +
    theme_bw(base_size = 30) +
    guides(
      col  = guide_legend(order = 1),
      fill = guide_legend(order = 2)
    )
}
```

```r
data2 = data2[data2['total_cells'] != 0]
metadata <- seurat_obj@meta.data

# Combine expression data2 with metadata2
data3 <- cbind(metadata, expr_matrix)
data3 = subset(data3, 'total_cells' != 0)
```

```r
# Function to run glmmTMB for single gene
run_glmmTMB <- function(gene) {
  print(gene)
  formula <- as.formula(paste(gene, "~ pct_SEAAD + pct_Mathys + pct_Gaz  + pseudotime_rank + offset(log_counts)"))
  model <- glmmTMB(formula, data = data3, family = nbinom2, ,
                    control = glmmTMBControl(parallel =  min(parallel::detectCores(),60)))
  return(model)
}
```

```r

results_regress <- list()

# Run glmmTMB for each gene 
for (gene in colnames(expr_matrix)) {
  print(gene)

  model <- run_glmmTMB(gene)
  print('model run')
  summary_model <- summary(model)

    coef_name <- 'pseudotime_rank'
    
      logFC <- summary_model$coefficients$cond[coef_name, "Estimate"]
      p_value <- summary_model$coefficients$cond[coef_name, "Pr(>|z|)"]
    
    results_regress[["pseudotime_rank"]] <- rbind(results_regress[["pseudotime_rank"]], data.frame(
      Gene = gene,
      LogFC = logFC,
      p_value = p_value
    ))
    
        
        
}
```

```r
res$p_adj = p.adjust(res$p_value, method = "BH")
res = results_regress$pseudotime_rank
res = subset(res, !is.na(p_value))
```

```r
# Log-transform the raw counts (with pseudocount)
log_counts <- log2(expr_matrix + 1)

# Calculate mean log2 expression for each gene
mean_expr <- colMeans(log_counts)

```

```r
df_labeled <- res2 %>%
      mutate(expr_bin = cut(MeanExpr, breaks = 7)) %>%
      group_by(expr_bin) %>%
      group_modify(~ {
    bind_rows(
      slice_max(.x, LogFC, n = 4, with_ties = FALSE),
      slice_min(.x, LogFC, n = 4, with_ties = FALSE)
    )
  }) %>%
  ungroup()


library(ggrepel)
library(ggplot2)

# Function to create an MA plot
MAPlot_nebula <- function(markers, colors, mean_expr = 0) {
    markers$X = markers$Gene  
    

    
    
    markers_to_label <- markers[(((markers$p_adj < 0.001 & abs(markers$LogFC_scaled) > 0.01) &  
                                  (markers$X %in% combined_nec & markers$MeanExpr > mean_expr)) | 
                                (markers$X %in% df_labeled$Gene)), ]

    
    
    # Print the number of genes that pass the label condition
    print(nrow(markers_to_label))  # Check how many labels will be plotted

    # Return the plot
    return(ggplot(markers, aes(x = MeanExpr, y = LogFC_scaled, 
                               fill = factor(ifelse(LogFC_scaled < 0 & p_adj < 0.05, "Down-regulated", 
                                    ifelse(LogFC_scaled > 0 & p_adj < 0.05, "Up-regulated", "NS"))))) +  
        
        # Points with colors indicating direction
        geom_point(size = 4, shape = 21, stroke = 0.1, color = "grey2", alpha = 0.8) +
        
        # Add horizontal line at y = 0 to represent no fold change
        geom_hline(yintercept = 0, color = "black", linetype = "dashed") +

        # Set plot theme
        theme_bw(base_size = 37) + 
        
        # Add labels for significant points (only for those passing the filter)
        geom_label_repel(
            data = markers_to_label,  # Only use the filtered data for labels
            aes(label = X),  # Label with the Gene name
            fill = alpha("white", 0.65), 
            color = "black",             
            size = 6,  # Reasonable size for labels
            max.overlaps = Inf,  # Limit label overlaps to avoid clutter
            force = 0.3 , 
            segment.size = 0.8 
        ) + 
        
        # Customize the color scale for the legend
        scale_fill_manual(name = "Direction (padj < 0.05)", values = colors) +
        
        # Customize x and y labels
        xlab("MeanExpr") + 
        ylab("Log2 Fold Change") +
        
        # Customize legends and guide
        guides(col = guide_legend(order = 1), fill = guide_legend(order = 2)))
    }

# Color palette (example)
cbPalette <- c("#3B4CC0", "grey", "#B40426", "#F0E442", "#0072B2", "#D55E00", "#CC79A7")
```

```r
MAPlot_nebula_markers <- function(markers, colors, mark = mark, mean_expr = 6) {

    # Add gene name column
    markers$X <- markers$Gene  

    # Define direction once
    markers$Direction <- factor(
        ifelse(markers$LogFC_scaled < 0 & markers$p_adj < 0.05, "Down-regulated",
               ifelse(markers$LogFC_scaled > 0 & markers$p_adj < 0.05, "Up-regulated", "NS"))
    )

    # Subset genes to label
    markers_to_label <- markers[
        markers$p_adj < 0.05 &
        markers$X %in% mark &
        markers$MeanExpr > mean_expr&
        abs(markers$LogFC_scaled) > 1.5,

        
    ]

    print(nrow(markers_to_label))

    ggplot(markers, aes(x = MeanExpr, y = LogFC_scaled, fill = Direction)) +  
        
        geom_point(size = 4, shape = 21, stroke = 0.1, color = "grey2", alpha = 0.8) +
        
        geom_hline(yintercept = 0, color = "black", linetype = "dashed") +

        theme_bw(base_size = 37) + 
        
        geom_label_repel(
            data = markers_to_label,
            aes(label = X, color = Direction),
            fill = alpha("white", 0.65), 
            size = 6,
            max.overlaps = Inf,
            force = 0.3, 
            segment.size = 0.8
        ) + 
        
        scale_fill_manual(name = "Direction (padj < 0.05)", values = colors) +
        #scale_color_manual(name = "Direction (padj < 0.05)", values = c(colors[0],colors[2],colors[1]) ) +
        #scale_color_manual(name = "Direction (padj < 0.05)", values = c(colors[1],colors[0],colors[2]) ) +
        scale_color_manual(name = "Direction (padj < 0.05)", values = c(colors[1],"red","orange"))  +

        xlab("MeanExpr") + 
        ylab("Log2 Fold Change") +
        
        guides(col = guide_legend(order = 1), fill = guide_legend(order = 2))
}
```

```r


res$z_LogFC <- (res$LogFC - mean(res$LogFC, na.rm = TRUE)) 

df = res
original_gene_list <- df$z_LogFC  
names(original_gene_list) <- df$Gene
gene_list<-na.omit(original_gene_list)
gene_list = sort(gene_list, decreasing = TRUE)
    
    
gse3 <- gseGO(geneList=gene_list, eps = 0,
             ont ="BP", 
             keyType = "SYMBOL", 
             minGSSize = 100, 
             maxGSSize = 1200, 
             pvalueCutoff = 0.05,
#              qvalueCutoff = 0.05,
             verbose = FALSE, 
             OrgDb = org.Hs.eg.db, 
             pAdjustMethod = "BH", 
             nPermSimple = 10000)
require(DOSE)

options(repr.plot.width=15, repr.plot.height=20)

clusterProfiler::dotplot(gse3, showCategory=60,x = "NES", split=".sign") + facet_grid(.~.sign) +
        ggtitle("ExN L2/3: metacells wrt PT rank") + theme_bw(base_size = 18) + 
          scale_y_discrete(labels=function(x) str_wrap(x, width=50))+
    theme(plot.margin = unit(c(1,1,1,3), "cm")) 
```

```r
library(clusterProfiler)
library(DOSE)

# Simplify (remove redundant GO terms)
gse3_simplified <- simplify(
  gse3,
  cutoff = 0.6,            # similarity threshold (0–1); smaller = more aggressive reduction
  by = "p.adjust",         
  select_fun = min,
  measure = "Wang"        
)

options(repr.plot.width=15, repr.plot.height=18)

clusterProfiler::dotplot(gse3_simplified, showCategory=50,x = "NES", split=".sign") + facet_grid(.~.sign) +
        ggtitle("ExN L2/3: metacells wrt PT rank") + theme_bw(base_size = 18) + 
          scale_y_discrete(labels=function(x) str_wrap(x, width=50))+
    theme(plot.margin = unit(c(1,1,1,3), "cm")) 
```

```r

options(repr.plot.width=18, repr.plot.height=10)

gse_simp_df = gse3_simplified@result
gse_simp_df <- gse_simp_df %>%
  mutate(.sign = ifelse(NES > 0, "Upregulated", "Downregulated"))

# Filter
gse_simp_df = subset(gse_simp_df, p.adjust < 0.05)

# Select top 20 per direction
gse_simp_df_head <- gse_simp_df %>%
  group_by(.sign) %>%
  arrange(qvalue, .by_group = TRUE) %>%
  slice_head(n = 18) %>% 
  arrange(NES, .by_group = TRUE) %>%
  mutate(Description = factor(Description, levels = unique(Description)))

# NES axis limits
max_abs_nes <- max(abs(gse_simp_df_head$NES), na.rm = TRUE)
min_abs_nes <- min(gse_simp_df_head$NES, na.rm = TRUE)

# New fill variable: log10 significance
gse_simp_df_head$log10q <- -log10(gse_simp_df_head$qvalue)

ggplot(gse_simp_df_head, aes(x = NES, y = Description, fill = log10q)) +
  geom_col(width = 0.65, color = "grey20", linewidth = 0.2) +
  geom_vline(xintercept = 0, color = "black", linewidth = 0.8) +

  # Sequential color scale for significance
  scale_fill_gradient(
    low = "#3B4CC0",    # blue (less significant)
    high = "#B40426",   # red (high significance)
    name = "-log10(q-value)"
  ) +

  coord_cartesian(xlim = c(min_abs_nes, max_abs_nes), expand = FALSE) +
  labs(
    title = "",
    x = "Normalized Enrichment Score (NES)",
    y = NULL
  ) +
  theme_minimal(base_size = 24, base_family = "Helvetica") +
  theme(
    plot.title = element_text(face = "bold", size = 24, hjust = 1, margin = margin(b = 15)),
    axis.text.y = element_text(size = 22, color = "black", margin = margin(r = 5)),
    axis.text.x = element_text(size = 20, color = "black"),
    axis.title.x = element_text(size = 20,),
    legend.position = "right",
    legend.title = element_text(size = 20, face = "bold"),
    legend.text = element_text(size = 17),
    panel.grid.major.y = element_blank(),
    plot.margin = unit(c(1, 1, 1, 3.5), "cm")
  ) +
  scale_y_discrete(labels = function(x) str_wrap(x, width = 60))

ggsave("./GO_metacells_pseudotime.pdf", plot = last_plot(), width = 25, height = 12)
```
